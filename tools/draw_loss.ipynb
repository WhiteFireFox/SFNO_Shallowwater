```python
import matplotlib.pyplot as plt

file_1_path = "../sfno_implement/script/sfno_loss_128.txt"
file_2_path = "../sfno_implement/script/sfno_loss_128_nonlinear.txt"
file_3_path = "../baseline_implement/sfno_torch_harmonics/script/sfno_loss_128.txt"
file_4_path = "../baseline_implement/sfno_torch_harmonics/script/sfno_loss_128_nonlinear.txt"
file_5_path = "../baseline_implement/fno_neuralop/script/sfno_loss_128.txt"
file_6_path = "../baseline_implement/fno_neuralop/script/sfno_loss_128_nonlinear.txt"

# Function to read data from the file and return steps and losses
def read_loss_data(file_path):
    steps = []
    losses = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split(',')
            step = int(parts[0].split()[1])
            loss = float(parts[1].split()[1])
            steps.append(step)
            losses.append(loss)
    return steps, losses

# Read data from both files
steps_1, losses_1 = read_loss_data(file_1_path)
steps_2, losses_2 = read_loss_data(file_2_path)
steps_3, losses_3 = read_loss_data(file_3_path)
steps_4, losses_4 = read_loss_data(file_4_path)
steps_5, losses_5 = read_loss_data(file_5_path)
steps_6, losses_6 = read_loss_data(file_6_path)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(steps_1, losses_1, label='SFNO (Ours), Linear')
plt.plot(steps_2, losses_2, label='SFNO (Ours), Non-Linear')
plt.plot(steps_3, losses_3, label='SFNO (TH), Linear')
plt.plot(steps_4, losses_4, label='SFNO (TH), Non-Linear')
plt.plot(steps_5, losses_5, label='FNO (NOP), Linear')
plt.plot(steps_6, losses_6, label='FNO (NOP), Non-Linear')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('L2 Loss Over Time (128 * 256)')
plt.legend()
plt.grid(True)
plt.show()
```
